ここの
### \begin{eqnarray}g_h(～)\end{eqnarray}\begin{eqnarray}g_y(～)\end{eqnarray}

は活性化関数であり、RNNの場合は一般的にghはrelu , gyは勾配が消失しやすいため、tanh（ハイポリックタンジェント）が使用されるそうです。

ということで実装してみます。

```python
#xtを作る。

import numpy as np

#繰り返し関数の定義

n =100
x = np.arange(n)
W1= np.arange(n)
Wy=[]
Wr=[]
bh = []
by = []

print(W1*x)

#ReLUの定義
def relu(a):
    return np.maximum(0,a)

#再帰により前の重みを計算する

def feedforward(n):
   
    def ht(n):
        if n == 2:
            return np.zeros(n)
        return relu(W1*x[n-1]+Wr*ht(n-2) +bh)

    return np.tanh(Wy*ht(n)+by)

feedforward(99)
```
